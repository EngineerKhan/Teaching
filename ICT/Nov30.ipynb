## Flights CSV

In Linear Regression, I mentioned a flights distance/time CSV. Its extremely easy to open it in Python (Pandas). You have to see it to believe it.

```python
!pip install pandas matplotlib
```

```python
import pandas as pd

flightsData = pd.read_csv("flightsDistanceDuration.csv")
```

And you are done. Now you can see the whole file.

```python
flightsData
```

Pandas library has its whole science and you would enjoy exploring it. We can also draw excellent graphs in Python using **Matplotlib**.

```python
import matplotlib.pyplot as plt

x = flightsData["distance"]
y = flightsData["duration"]

plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.xlabel("column1")
plt.ylabel("column2")
plt.title("Graph of Distance vs Time")
plt.show()
```
